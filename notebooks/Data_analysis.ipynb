```python
import pandas as pd
import sys
sys.path.insert(0, "src")
from src.data.format_prompts import build_prompt, build_full_text, COMPLETION_TEMPLATE

train = pd.read_parquet("data/processed/train.parquet")
val   = pd.read_parquet("data/processed/val.parquet")

def show(df, name, n=3):
    print("=" * 70)
    print(f"{name} — {len(df):,} rows")
    print("=" * 70)
    for row in df.sample(n, random_state=42).to_dict(orient="records"):
        print(f"INPUT:\n{build_prompt(row)}")
        print(f"EXPECTED OUTPUT:\n{COMPLETION_TEMPLATE.format(priority=row['priority'], team=row['team'])}")
        print(f"FULL STRING:\n{build_full_text(row)}")
        print("-" * 70)

show(train, "TRAIN")
show(val, "VAL")
```
